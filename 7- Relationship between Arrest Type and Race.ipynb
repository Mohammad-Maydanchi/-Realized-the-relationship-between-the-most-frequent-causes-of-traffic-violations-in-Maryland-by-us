```python
###For simplicity we devide our numbers to 1000 and round them.
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# sphinx_gallery_thumbnail_number = 2

Race = ["ASIAN", "BLACK", "HISPANIC", "NATIVE AMERICAN",
              "WHITE", "OTHER"]
ArrestType = ["A - Marked Patrol", "B - Unmarked Patrol", "L - Motorcycle",
           "O - Foot Patrol", "Q - Marked Laser", "S - License Plate Recognition"]

numbers = np.array([[50,2, 1, 1, 7, 0.3],
                    [285,10, 2, 3, 24, 44],
                    [199, 6, 2, 2, 2, 2],
                    [284, 0.06, 0, 0, 0.08, 0],
                    [28, 13, 4, 4, 45, 4],
                    [46, 20, 0.4, 0.4, 6, 0.6]])


fig, ax = plt.subplots()
im = ax.imshow(numbers)

# We want to show all ticks...
ax.set_xticks(np.arange(len(ArrestType)))
ax.set_yticks(np.arange(len(Race)))
# ... and label them with the respective list entries
ax.set_xticklabels(ArrestType)
ax.set_yticklabels(Race)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(Race)):
    for j in range(len(ArrestType)):
        text = ax.text(j, i, numbers[i, j],
                       ha="center", va="center", color="b")

ax.set_title("Relationship between Arrest Type and Race")
fig.tight_layout()
plt.show()
```
