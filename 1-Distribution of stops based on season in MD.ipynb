# 1- Which season has most stops in Maryland?

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv("E:/Education/Auburn/2018-Fall/Info-Jeff/project/Traffic_Violations.csv",parse_dates=['Date'])
```

```python
df.head(5)
```

```python
df2=df[['State','Date',]]
df2.head(10)
```

# Maryland

```python
####### the State which has  most Stops
df5=df2.groupby(['State']).State.value_counts().nlargest(1)
df5
```

```python
df8=df2.loc[(df2['State'] == 'MD')]
df8
```

```python
###the number of stops for Maryland during winter
mask111 = ((df8['Date'] >= '2018-01-01') & (df8['Date'] < '2018-04-01')
         |(df8['Date'] >= '2017-01-01') & (df8['Date'] < '2017-04-01')
         |(df8['Date'] >= '2016-01-01') & (df8['Date'] < '2016-04-01')
         |(df8['Date'] >= '2015-01-01') & (df8['Date'] < '2015-04-01')
         |(df8['Date'] >= '2014-01-01') & (df8['Date'] < '2014-04-01')
         |(df8['Date'] >= '2013-01-01') & (df8['Date'] < '2013-04-01')
         |(df8['Date'] >= '2012-01-01') & (df8['Date'] < '2012-04-01'))
WinterMD=df8.loc[mask111]
WinterMD
```

```python
WMD=WinterMD.shape[0]
print(f'The number of stops for Maryland during winter is:{WMD}')
```

```python
###The number of stops for Maryland during spring
mask222 = ((df8['Date'] >= '2018-04-01') & (df8['Date'] < '2018-07-01')
         |(df8['Date'] >= '2017-04-01') & (df8['Date'] < '2017-07-01')
         |(df8['Date'] >= '2016-04-01') & (df8['Date'] < '2016-07-01')
         |(df8['Date'] >= '2015-04-01') & (df8['Date'] < '2015-07-01')
         |(df8['Date'] >= '2014-04-01') & (df8['Date'] < '2014-07-01')
         |(df8['Date'] >= '2013-04-01') & (df8['Date'] < '2013-07-01')
         |(df8['Date'] >= '2012-04-01') & (df8['Date'] < '2012-07-01'))
SpringMD=df8.loc[mask222]
SpringMD
```

```python
SpMD=SpringMD.shape[0]
print(f'The number of stops for Maryland during spring is:{SpMD}')
```

```python
###The number of stops for Maryland during summer
mask333 = ((df8['Date'] >= '2018-07-01') & (df8['Date'] < '2018-10-01')
         |(df8['Date'] >= '2017-07-01') & (df8['Date'] < '2017-10-01')
         |(df8['Date'] >= '2016-07-01') & (df8['Date'] < '2016-10-01')
         |(df8['Date'] >= '2015-07-01') & (df8['Date'] < '2015-10-01')
         |(df8['Date'] >= '2014-07-01') & (df8['Date'] < '2014-10-01')
         |(df8['Date'] >= '2013-07-01') & (df8['Date'] < '2013-10-01')
         |(df8['Date'] >= '2012-07-01') & (df8['Date'] < '2012-10-01'))
SummerMD=df8.loc[mask333]
SummerMD
```

```python
SuMD=SummerMD.shape[0]
print(f'The number of stops for Maryland during summer is:{SuMD}')
```

```python
###The number of stops for Maryland during fall
mask444 = ((df8['Date'] >= '2018-10-01') & (df8['Date'] <= '2018-12-31')
         |(df8['Date'] >= '2017-10-01') & (df8['Date'] <= '2017-12-31')
         |(df8['Date'] >= '2016-10-01') & (df8['Date'] <= '2016-12-31')
         |(df8['Date'] >= '2015-10-01') & (df8['Date'] <= '2015-12-31')
         |(df8['Date'] >= '2014-10-01') & (df8['Date'] <= '2014-12-31')
         |(df8['Date'] >= '2013-10-01') & (df8['Date'] <= '2013-12-31')
         |(df8['Date'] >= '2012-10-01') & (df8['Date'] <= '2012-12-31'))
FallMD=df8.loc[mask444]
FallMD
```

```python
FMD=FallMD.shape[0]
print(f'The number of stops for Maryland during fall is:{FMD}')
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = ('Winter', 'Spring', 'Summer', 'Fall')
y_pos = np.arange(len(objects))
performance = [WMD,SpMD,SuMD,FMD]
 
plt.bar(y_pos, performance, align='center', alpha=0.5,color='g')
plt.xticks(y_pos, objects)
plt.ylabel('Stops')
plt.title('Distribution of stops based on season in MD')
 
plt.show()
```
