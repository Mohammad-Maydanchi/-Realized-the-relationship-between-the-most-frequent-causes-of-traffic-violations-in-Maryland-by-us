# 5-Do drivers who are living in a specific state and have gotten driving license from there, care about rules more than people who have gotten driving license from other states and they are just traveling to that specific state?

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv("E:/Education/Auburn/2018-Fall/Info-Jeff/project/Traffic_Violations.csv")
```

```python
df2=df[['State','DLState']]
df2
```

```python
###Stops for Host Drivers
mask1 = (df2['State'] ==df2['DLState'])
       
H=df2.loc[mask1]
Hs=H.count()/df2.count()
Hs
```

```python
###Stops for Host Drivers
mask1 = (df2['State'] !=df2['DLState'])
       
G=df2.loc[mask1]
Gu=G.count()/df2.count()
Gu
```

```python
import numpy as np
import matplotlib.pyplot as plt


N = 1
Host = (100*Hs)
Gust = (100*Gu)

ind = np.arange(N)    # the x locations for the groups
width = 0       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, Host, width=0.01)
p2 = plt.bar(ind, Gust, width=0.01,
             bottom=Host)

#plt.ylabel('Scores')
plt.title('Host drivers stops vs guest drivers stops')
plt.xticks(ind, ('G1'))
plt.yticks(np.arange(0, 91, 10))
plt.legend((p1[0], p2[0]), ('Host', 'Guest'))

plt.show()
```

# Conclusion

Based on this information we can say drivers who come from other state care more about rules. But, this information is not enough because most of transportations in a specific state are done with people who are living in that state.
